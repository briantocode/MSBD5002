# Define function

```python

def str_to_ts(x):
    x_ts = datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S")
    return x_ts
def get_mean(x,y):
    if np.isnan(x):
        return y
    elif np.isnan(y):
        return x
    else:
        return (x+y)/2
```

```python
def fill_num_na_closetime(df):
    df[num_features] = df[num_features].fillna(method='pad',limit = 5)
    return df
def fill_cat_na(df):
    df[cat_feature] = df[cat_feature].fillna(df.mode().iloc[0])
    return df
def fill_num_na_mean(df):
    df[num_features] = df[num_features].fillna(df.mean())
    return df
```

```python
train_file = '../from_Brian/_myTRAIN.csv'
train_df =pd.read_csv(train_file,header=[0,1], parse_dates = True, index_col = 0)
```

# change the structure of the data

```python
'''
transfer timestamp to number
after this , the time is 0~10000+
if want to rerun this part, it should be together with .read_csv
'''
train_df = train_df.reset_index().reset_index()
```

```python
for Id in stationId:
    train_df[Id, 'time'] = train_df['index']
train_df = train_df.drop('index', axis = 1,level = 0)
```

```python
'''
restucture the data, now they are in station list,use the last line to concat them
use list to process them seperatelly
'''

station_list = []
timelen = 11618
for Id in stationId:
    station_df = train_df[Id].copy()
    station_df['station_id'] = [Id]*timelen
    station_df = station_df.reset_index()
    station_list.append(station_df)
```

#  generate the final weather feature

```python
'''
after this prograss
we have a list to store 35 stations with aq and weather feature

'''
for i in range(len(station_list)):
    station_df = station_list[i]
    for feature in weatherFeatures:
            gr_feature = feature[0]
            ob_feature = feature[1]
            mean_feature = feature[2]
            station_df[mean_feature] = station_df.apply(lambda x:get_mean(x[gr_feature],x[ob_feature]),  axis=1)
    station_list[i] = station_df.drop(features_to_drop,axis = 1)
    
```

```python
'''
store each element in the list to csv
'''
for i in range(len(station_list)):
    temp = station_list[i]
    temp.to_csv('../data/train_by_station/no_fill_missing_value/'+temp.loc[1,'station_id']+'.csv')
```

```python
all_sta_df = pd.concat(station_list)
```

```python
'''
store the big table to csv
'''
all_sta_df.to_csv('../data/train_by_station/no_fill_missing_value/allstations.csv')
```

# fill missing value

## same time, fill it with mean of the other station

```python
all_sta_df_time = all_sta_df.sort_values(['time'])
```

```python
all_sta_df_time.set_index(['time','station_id'],inplace = True)
```

```python
all_sta_df_time.isna().sum()
```

```python
''' For the missing value,at the same time if the missing is less than 24, fill it by mean of the dataset

and for missing value more than this, fill it with

'''
time_index_list = list(range(11618))
for i in range(11618):
    df = all_sta_df_time.loc[i]
    for feature_ in num_features:
        b = df[feature_].isna().sum()
        if b<24:
            df[feature_]= df[feature_].fillna(df[feature_].mean())
    for feature_ in cat_feature:
        b = df[feature_].isna().sum()
        if b<24:
            df[feature_]= df[feature_].fillna(df[feature_].mode())
```
